## Pregunta 1

Definir una función max_de_dos(), que tome dos números como argumentos y devuelva el mayor de ellos. Si los números son iguales que imprima en pantalla "Los números son iguales" y no regrese ningún valor.


```
# Escriba aquí su código
def max_de_dos():
  a = input('Ingrese numero 1: ')
  b = input('Ingrese numero 2: ')
  if a > b:
    print('El mayor es: ', a)
  if a < b:
    print('El mayor es: ', b)
  if a==b:
    print('Los numeros son iguales.')

max_de_dos()
```

## Pregunta 4

Juguemos al juego de adivinar el numero, generaremos un número entre 1 y 20.

Nuestro objetivo es adivinar el número. Si fallamos nos dirán si es mayor o menor que el número buscado. También poner el número de intentos requeridos.

```
# Escriba aquí su código
import random

a = random.randint(1,20)

b = int(input('Ingrese un numero para empezar: '))
print('')

intentos = 0

while a != b:
  if b > a:
    print('El numero es grande.')
    b = int(input('Ingrese un nuevo numero: '))
    print('')
    intentos += 1

  if b < a:
    print('El numero es pequenio.')
    b = int(input('Ingrese un nuevo numero: '))
    print('')
    intentos += 1

if a == b:
  print('Numero aleatorio: ', a)
  print('Numero escrito: ', b)
  print('Numero de intentos: ', intentos)
```

## Pregunta 6

Defina una función llamada `es_numero()` que evalúa si la entrada es un número y que retorne una variable booleana según sea el caso

```
# Escriba aquí su código

def es_numero(a):
  if a.isnumeric():
    print('Es numero.')
  else:
    print('NO es numero')

a = input('Ingrese numero: ')

es_numero(a)
```

## Pregunta 7

* El nombre y la ubicación deben ser cadenas no vacías. 
* El puntaje debe ser un número (sin restricciones sobre su valor) 
* El precio debe ser un número distinto de cero.

si no se cumple alguna de las condiciones, ejecute un error que muestre en pantalla el problema.

*Pista*: Use las funciónes `es_cadena_no_vacia()` y `es_numero()`




```
# Escriba aquí su código

class Hotel():
  nombre = ''
  ubicacion = ''
  puntaje = ''
  precio = ''


hotel = Hotel()
hotel.nombre = input('Escriba el nombre del hotel: : ')
hotel.ubicacion = input('Escriba la ubicacion del hotel: ')
hotel.puntaje = input('Escriba el puntaje del hotel: ')
hotel.precio = input('Escriba el precio del hotel: ')

es_cadena_no_vacia(hotel.nombre)
es_cadena_no_vacia(hotel.ubicacion)

es_numero(hotel.puntaje)

if int(hotel.precio) != 0:
  es_numero(hotel.precio)
else:
  print('El precio NO puede contener un valor de 0')
```
