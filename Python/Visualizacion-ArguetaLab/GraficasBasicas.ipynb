```python
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
#Estilos disponibles.
```

```python
datos = [1, 3, 5, 7]

plt.style.use("dark_background")
fig, ax = plt.subplots() #Regresa una lista con el tamanio de la figura que tiene predeterminada 432x288
ax.scatter((2,1),(3,2), s =22)
ax.plot(datos, linewidth = 5) #Ingresa la informacion de manera entendible.
ax.set_title("Linea y puntos.", fontsize = 24)
ax.set_xlabel("valores", fontsize = 10)
ax.set_ylabel("datos", fontsize = 10)
plt.show()

#fig = figure(figura).
#ax = axes(ejes).
#figsize(x,y) = modifica el tamanio de la visualizacion (horizontal, vertical).
#linewidth = modifica el grosor de la linea plot.
#set_title = Agrega titulo a la visualizacion.
#set_xlabel = Agrega una descripcion al eje x.
#set_ylabel = Agrega una descripcion al eje x.
#fontsize = modifica el tamanio del texto.
#plt.style.use("Estilo") = Agrega un estilo prediseniado a la visualizacion

#plot = grafica de linea.
#scatter((x,y)) = grafica de puntos. s = Tamanio de los puntos.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.style.use("dark_background")
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Punto", fontsize = 24)
ax.set_xlabel("valores", fontsize = 10)
ax.set_ylabel("datos", fontsize = 10)
plt.show()
```
